```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Config de graficos
plt.style.use('seaborn-v0_8')
sns.set_palette('viridis')
%matplotlib inline
```

```python
#Descargar Data Set

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
df = pd.read_csv(url, sep=";")
df.head()
```

```python
#Dimensiones

print("Dimensiones del dataset:", df.shape)
#Informacion general
df.info()

#Estadisticas
df.describe()
```

```python
#Conteo de Clases
df['quality'].value_counts().sort_index()
```

```python
#Visual
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='quality')
plt.title('Calidad de Vino')
plt.xlabel('Calidad (0-10)')
plt.ylabel('Cantidad')
plt.show()
```

```python
#Matriz Correlacion
correlation_matrix = df.corr(numeric_only=True)

#Filtro de Correlaciones
cor_quality = correlation_matrix['quality'].sort_values(ascending=False)

print('Correlaciones con ', 'quality')
print(cor_quality)
```

```python
#Grafico de calor
plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Matriz de Correlacin')
plt.show()
```

```python
#Relaciones con variables mas importanes (Alchol, volatile acidity, sulphates)

fig, axs = plt.subplots(1, 3, figsize=(18, 5))

sns.boxplot(data=df, x='quality', y='alcohol', ax=axs[0])
axs[0].set_title('Alcohol vs Calidad')

sns.boxplot(data=df, x='quality', y='volatile acidity', ax=axs[1])
axs[1].set_title('Volatilidad Acida vs Calidad')

sns.boxplot(data=df, x='quality', y='sulphates', ax=axs[2])
axs[2].set_title('Sulfatos vs Calidad')

plt.tight_layout()
plt.show()
```

```python
#Verificar outliners y valores nulos

df.plot(kind='box', subplots=True, layout=(4,3), figsize=(15,10) , sharex=False , sharey=False)
plt.tight_layout()
plt.show()

#Valores Null
df.isnull().sum()
```
